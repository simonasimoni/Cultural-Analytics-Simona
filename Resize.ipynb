# Die Größe der Bilder anpassen

```python
import os
import cv2
```

```python
def resize_and_save_images(input_folder, output_folder, target_size):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        if filename.endswith(('.jpg', '.jpeg', '.png')):
            image_path = os.path.join(input_folder, filename)
            image = cv2.imread(image_path)

            resized_image = cv2.resize(image, target_size, interpolation=cv2.INTER_LINEAR)
            output_path = os.path.join(output_folder, filename)
            cv2.imwrite(output_path, resized_image)
            print(f"Die Bilder wurden auf: {output_path} gespeichert")
```

```python
input_folder = "/home/sc.uni-leipzig.de/lj25vapa/Cultural Analytics/Alle_Bilder/Nordkorea"
output_folder = "/home/sc.uni-leipzig.de/lj25vapa/Cultural Analytics/DPRK"

resize_and_save_images(input_folder, output_folder, target_size=(640 , 640))
```
